```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# Set up the Selenium WebDriver (make sure you have the correct driver installed)
driver = webdriver.Chrome()  # or use webdriver.Firefox() for Firefox

try:
    # Open the EPA EIS search page
    driver.get("https://cdxapps.epa.gov/cdx-enepa-II/public/action/eis/search")

    # Fill in search criteria (this may vary based on the actual input fields)
    start_date = driver.find_element(By.NAME, "searchCriteria.startFRDate")
    start_date.clear()
    start_date.send_keys("01/01/2024")

    end_date = driver.find_element(By.NAME, "searchCriteria.endFRDate")
    end_date.clear()
    end_date.send_keys("12/31/2024")

    # Submit the form (you might need to locate the submit button and click it)
    search_button = driver.find_element(By.XPATH, "//button[contains(text(), 'Search')]")
    search_button.click()

    time.sleep(5)  # Wait for the results to load

    # Collect results (adjust selectors based on the actual result structure)
    records = driver.find_elements(By.CSS_SELECTOR, ".record")
    
    observations = []
    for record in records:
        title = record.find_element(By.TAG_NAME, "h3").text
        date = record.find_element(By.CSS_SELECTOR, ".date").text
        observations.append({"title": title, "date": date})

    # Output results
    for obs in observations:
        print(f"Title: {obs['title']}, Date: {obs['date']}")

finally:
    driver.quit()  # Ensure the browser is closed after the script runs
```
